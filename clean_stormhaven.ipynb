Compute the mean and standard deviation of housing prices in each zip code and city

Check how many entries have Nan for zipcode or city.

```python
# Count entries with NaN in zip_code
nan_in_zipcode = housing_df_cleaned['zip_code'].isna().sum()
print(f"Number of entries with NaN in zip_code: {nan_in_zipcode}")

# Count entries with NaN in city
nan_in_city = housing_df_cleaned['city'].isna().sum()
print(f"Number of entries with NaN in city: {nan_in_city}")

# Count entries with NaN in both zip_code and city
nan_in_both = housing_df_cleaned[housing_df_cleaned['zip_code'].isna() & housing_df_cleaned['city'].isna()].shape[0]
print(f"Number of entries with NaN in both zip_code and city: {nan_in_both}")
```

We can drop these entries because our original df has over 2 mil entries.

```python
# Drop rows where either zip_code or city has NaN values
housing_df_cleaned = housing_df_cleaned.dropna(subset=['zip_code', 'city'])
```

```python
# Group by 'zip_code' and 'city' and compute mean and std for 'price'
grouped_stats = housing_df_cleaned.groupby(['zip_code', 'city'])['price'].agg(['mean', 'std']).reset_index()

# Rename the columns for clarity
grouped_stats.rename(columns={'mean': 'price_mean', 'std': 'price_std'}, inplace=True)

# Merge the statistics back into the original DataFrame
housing_df_cleaned = housing_df_cleaned.merge(grouped_stats, on=['zip_code', 'city'], how='left')
```

```python
housing_df_cleaned.head(5)
```

Calculate the price per square foot, price per acre of land, bedrooms to bathrooms ratio, price per bedroom, has acreage (has any lawnspace or not)



```python
# Add a column for price per square foot
housing_df_cleaned['price_per_sqft'] = housing_df_cleaned['price'] / housing_df_cleaned['house_size']

# Add a column for price per acre of land
housing_df_cleaned['price_per_acre'] = housing_df_cleaned['price'] / housing_df_cleaned['acre_lot']

# Add a column for bedrooms to bathrooms ratio
housing_df_cleaned['bed_bath_ratio'] = housing_df_cleaned['bed'] / housing_df_cleaned['bath']

# Add a column for price per bedroom
housing_df_cleaned['price_per_bedroom'] = housing_df_cleaned['price'] / housing_df_cleaned['bed']

# Add a column to indicate if the property has any lawn space
housing_df_cleaned['has_acreage'] = housing_df_cleaned['acre_lot'] > 0
```
